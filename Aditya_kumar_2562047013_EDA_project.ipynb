 Hungry Hours: Analyzing Peak Ordering Trends and Customer Preferences

First we will check Peak Ordering Trends (Meal Category)

```python
print("=== Peak Ordering Trends (Meal Category) ===")
meal_counts = df['Frequently ordered Meal category'].value_counts()
print(meal_counts)
print("\n")
```

Now we will plot our first graph for First we will check Peak Ordering Trends (Meal Category)






```python

plt.figure(figsize=(10, 6))
plt.bar(meal_counts.index, meal_counts.values)
plt.xlabel('Meal Category')
plt.ylabel('Number of Orders')
plt.title('Peak Ordering Trends (Meal Category)')
```

Now we will check for  Customer Preferences (Veg / Non-Veg)

```python
print("=== Customer Preferences (Veg / Non-Veg) ===")
pref_counts = df['Perference'].value_counts()
print(pref_counts)
print("\n")
```

Now will add a bar graph for understanding this data

```python
plt.figure(figsize=(10, 6))
plt.bar(pref_counts.index, pref_counts.values)
plt.xlabel('Preference')
plt.ylabel('Number of Customers')
plt.title('Customer Preferences (Veg / Non-Veg)')
```

Now we will check for maximum waiting time prerence


```python
print("=== Maximum Wait Time Preference ===")
wait_counts = df['Maximum wait time'].value_counts()
print(wait_counts)
print("\n")
```

NOW we will add one more graph *italicized text*

```python
plt.figure(figsize=(10, 6))
plt.bar(wait_counts.index, wait_counts.values)
plt.xlabel('Wait Time')
plt.ylabel('Number of Customers')
plt.title('Maximum Wait Time Preference')
```

Now in last we will check for impact of rating on delivery

```python
print("=== Influence of Rating ===")
rating_influence = df['Influence of rating'].value_counts()
print(rating_influence)
print("\n")
```

Now we will plot a pie chart for understanding for this data

```python
plt.figure(figsize=(8, 8))
plt.pie(rating_influence.values, labels=rating_influence.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Influence of Rating')
#
```

### Project Observations:

Based on the analysis, here are some key observations:


#### Peak Ordering Trends (Meal Category):
*   **Lunch** is the most frequently ordered meal category, followed by **Snacks**.
*   **Dinner** and **Breakfast** have fewer orders compared to Lunch and Snacks.

#### Customer Preferences (Veg / Non-Veg):
*   A significant majority of customers prefer **Non-Veg foods** (Lunch / Dinner).
*   **Veg foods** are also popular, but to a lesser extent.
*   **Sweets** and **Bakery items** have very low preference rates, suggesting they are niche categories or not primarily ordered via these services.

#### Maximum Wait Time Preference:
*   Customers show a strong preference for wait times of **30 minutes** and **45 minutes**, with 45 minutes being slightly more preferred.
*   Fewer customers are willing to wait for **60 minutes** or **More than 60 minutes**, indicating a desire for relatively quick delivery.

#### Influence of Rating:
*   A large majority of customers (**75.8%**) are influenced by ratings when making ordering decisions, highlighting the importance of positive restaurant and delivery ratings.
*   A smaller portion (**14.4%**) are not influenced, and some (**9.8%**) are ambivalent ('Maybe').
